Part 2 : features and label (x , y)

```python
y = dataset['Exited']
x = dataset[['CreditScore','Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
# x = dataset.drop([['RowNumber', 'CustomerId', 'Surname', 'Geography', 'Exited']])
```

```python
y.head()
```

# Part 3 : Pre-Processing

### Encoding Categorical data

```python
male = pd.get_dummies(x['Gender'], drop_first=True)
male
```

```python
x = pd.concat([x , male], axis=1)
x
```

```python
x.drop(['Gender'], axis=1, inplace=True)
x
```

# Part 4 : Train the model (model selection)

### train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=101)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Model: Multy-Layer Perceptron

```python
from keras.models import Sequential
```

```python
mlp = Sequential()
```

```python
from keras import layers
```

```python
mlp.add(layers.Dense(units=6, activation='relu'))
mlp.add(layers.Dense(units=6, activation='relu'))
mlp.add(layers.Dense(units=1, activation='sigmoid'))
```

# Part 5 : Training the model

```python
mlp.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
mlp.fit(x_train, y_train, batch_size=32, epochs=100)
```

# Part 6: Evaluation of model and prediction

```python
x.head()
```

```python
pred = mlp.predict(x_test)
```

```python
pred_df = pd.DataFrame(data = pred)
pred_df
```

```python
pred_df = pred_df > 0.5
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, pred_df))
```

```python
print(classification_report(y_test, pred_df))
```
